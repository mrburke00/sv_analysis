```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sv_info_file = 'horse/akh_ins_info.txt'
output_file = 'horse/akh_insertion_246.txt'
```

```python
#INSERTIONS#
df = pd.read_csv(sv_info_file, sep='\t', lineterminator='\n')
print(df.head())
#df.columns = ['chrom','start','end','length']
df.columns = ['chrom','start','end','length','seq']
print(df.head())
```

```python
#df_new = df.loc[df['length'] == '246']
df_new = df[(df['length'] >= '244') & (df['length'] <= '246')]
```

```python
print(len(df_new))
df_new = df_new.drop_duplicates(subset='start', keep='first')
print(len(df_new))
```

```python
with open(output_file, 'w') as f:
    for i,row in df_new.iterrows():
        coords = '>'+str(row['chrom'])+':'+str(row['start'])+'-'+str(row['end'])
        #coords = str(row['chrom'])+'\t'+str(row['start'])+'\t'+str(row['end'])
        f.write(coords)
        f.write('\n')
        f.write(row['seq'])
        f.write('\n')
```

```python
print('done')
```
