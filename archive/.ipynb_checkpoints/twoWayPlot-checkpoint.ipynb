```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Hanoverian', 'Przewalski', 'Thoroughbred', 'Jeju', 'Curly', 'Fresian', 'Debao', 'Criollo', 'Mongolian', 'American Quarter', 'Arabian', 'Akhal-Teke'
sizes = [5, 5, 10, 5, 5, 20, 10, 5, 5, 5, 10, 10]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Species Distribution")
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'European', 'Mongolian', 'Korean', 'North American', 'Chinese', 'South American','Middle Eastern'

sizes = [3, 2, 1, 2, 1, 1, 2]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Geographic Distribution")
plt.show()
```

```python
df_feral = pd.read_csv('horse/feral_out.txt', sep='\t', lineterminator='\n')
df_feral.columns = ['chrom','start','end','subfam','fam']
#print(df_feral.head())
df_thor = pd.read_csv('horse/thor_out.txt', sep='\t', lineterminator='\n')
df_thor.columns = ['chrom','start','end','subfam','fam']
```

```python
x_f = df_feral['subfam'].value_counts()
y_f = df_feral['fam'].value_counts()
```

```python
print(df_feral['subfam'].unique())
keys_f = df_feral['subfam'].unique()
keys_t = df_thor['subfam'].unique()
keys = list(keys_f) + list(keys_t)
```

```python
f_counts = {f:0 for f in keys}
t_counts = {f:0 for f in keys}
for i,row in df_feral.iterrows():
    f_counts[row['subfam']] -= 1
for i,row in df_thor.iterrows():
    t_counts[row['subfam']] += 1
```

```python
print(t_counts)
print(f_counts)
```

```python
from matplotlib.pyplot import figure



fig = plt.figure()
figure(figsize=(24,24), dpi=100)
plt.bar(t_counts.keys(), f_counts.values(), width=1, color='r', label = 'Przewalski')
plt.bar(t_counts.keys(), t_counts.values(), width=1, color='b', label = 'Thoroughbred')
plt.xticks(rotation = 65)
plt.legend()
plt.show()
```
